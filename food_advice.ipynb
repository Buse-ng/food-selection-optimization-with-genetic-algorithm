```python
target_calories = 2000
population_size = 100
mutation_prob = 0.1
```

```python
def fitness(individual):
    total_cost = sum(menu[meal]['ücret'] for meal in individual)
    total_calories = sum(menu[meal]['kalori'] for meal in individual)
    calorie_difference = abs(total_calories - target_calories)
    return 1 / (total_cost + calorie_difference)
```

```python
def create_population():
    population = []
    for _ in range(population_size):
        individual = random.sample(menu.keys(), random.randint(1, len(menu)))
        population.append(individual)
    return population
```

```python
def selection(population):
    tournament_size = 3
    tournament_indices = random.sample(range(len(population)), tournament_size)
    tournament_fitnesses = [fitness(population[i]) for i in tournament_indices]
    return population[tournament_indices[tournament_fitnesses.index(max(tournament_fitnesses))]]
```

```python
def crossover(parent1, parent2):
    point = random.randint(0, len(parent1)-1)
    child = parent1[:point] + parent2[point:]
    return child
```

```python
def mutation(individual):
    for i in range(len(individual)):
        if random.uniform(0, 1) < mutation_prob:
            j = random.randint(0, len(individual)-1)
            individual[i], individual[j] = individual[j], individual[i]
    return individual
```

```python
def create_new_population(population):
    new_population = []
    fitnesses = [fitness(individual) for individual in population]
    
    # En iyi 2 bireyi seç ve yeni popülasyona ekle
    
    best_individuals_indices = sorted(range(len(fitnesses)), key=lambda k: fitnesses[k], reverse=True)[0:4]
    for index in best_individuals_indices:
        new_population.append(population[index])

        
    for _ in range(population_size-4):
        parent1 = selection(population)
        parent2 = selection(population)
        child = crossover(parent1, parent2)
        child = mutation(child)
        new_population.append(child)
    return new_population
```

```python
def genetic_algorithm():
    population = create_population()
    for _ in range(100):
        population = create_new_population(population)
    best_individual = max(population, key=fitness)
    return best_individual
```

```python
for _ in range(10):
    best_set = genetic_algorithm()
    best_calories = sum(menu[meal]['kalori'] for meal in best_set)
    best_cost = sum(menu[meal]['ücret'] for meal in best_set)
    best_fit = fitness(best_set)

    print("En uygun ve en ucuz yemek listesi:")
    for meal in best_set:
        print("- " + meal)
        
    print("Toplam kalori:", best_calories)
    print("Toplam maliyet:", best_cost)
    print("Fitness Değeri: {:.4f}".format(best_fit))
    
    print("\n")
```
